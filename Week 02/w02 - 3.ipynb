```python
import random
#1과 6 사이의 정수 난수
print(random.randint(1,6))
#1과 6 사이(1을 포함하지 않음)의 정수 난수
print(random.randrange(1,6))
#0과 6사이(6을 포함하지 않음)의 정수 난수
print(random.randrange(6))
#리스트에서 무작위 요소 선택
my_list = [1,2,3,4,5]
print(random.choice(my_list))
#리스트 요소 무작위로 섞기
random.shuffle(my_list)
print(my_list)
```
