```python
celsius = int(input('섭씨 온도를 입력하세요 : '))
fahrenheit = (9/5) * celsius + 32
print('섭씨',celsius, '도는 화씨,',fahrenheit, '도 입니다')
```

```python
radius = int(input('원의 반지름을 입력하세요'))
pi = 3.141592
area = pi * radius ** 2
circumference = 2 * pi * radius
print('원의 둘레 =',str(circumference) + '원의 면적 =', area)
```
