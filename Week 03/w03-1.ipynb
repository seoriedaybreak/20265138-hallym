```python
weight=float(input('몸무게를 kg 단위로 입력:'))
height=float(input('키를 미터 단위로 입력'))
bmi=(weight/(height**2))
print(f'당신의 BMI={bmi:.2f}입니다.')
```

```python
#원금, 이자율, 기간 입력 받기
money=float(input('원금을 입력하세요:'))
rate=float(input('이자율을 입력하세요, 예: 0.05 for 5%'))
year=int(input('기간(년)을 입력하세요:'))

#원리금 계산
total=money*(1+rate)**year

#결과 출력
print(f'{year}년 후의 원리금은 {total:.2f}원 입니다')
```
